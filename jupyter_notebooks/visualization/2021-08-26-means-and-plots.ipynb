# Calculating mean and std deviation of accuracies 

```python
import pandas as pd
```

```python
outfile='/cluster/gjb_lab/2472402/results/25Aug_CV_summary.csv'
expt_name='25Aug'
with open(outfile, 'w+') as f:
    # write header
    f.write('fold,SNNS_HMM_acc,Keras_HMM_acc,SNNS_PSSM_acc,Keras_PSSM_acc,SNNS_HMM_sov,Keras_HMM_sov,SNNS_PSSM_sov,Keras_PSSM_sov\n')
    for i in range(1,8):
        keras_path="/cluster/gjb_lab/2472402/results/keras/%s/cv%d_scores.csv" % (expt_name,i)
        snns_path="/cluster/gjb_lab/2472402/results/snns/%s/cv%d_scores.csv" % (expt_name,i)
        snns=pd.read_csv(snns_path).set_index('seqID')
        keras=pd.read_csv(keras_path).set_index('seqID')
        f.write("%d," % i)
        f.write("%.2f ± %.2f" % (snns.HMM_acc.mean(),snns.HMM_acc.std()) + ',')
        f.write("%.2f ± %.2f" % (keras.HMM_acc.mean(),keras.HMM_acc.std()) + ',')
        f.write("%.2f ± %.2f" % (snns.PSSM_acc.mean(),snns.PSSM_acc.std()) + ',')
        f.write("%.2f ± %.2f" % (keras.PSSM_acc.mean(),keras.PSSM_acc.std()) + ',')
        f.write("%.2f ± %.2f" % (snns.HMM_sov.mean(),snns.HMM_sov.std()) + ',')
        f.write("%.2f ± %.2f" % (keras.HMM_sov.mean(),keras.HMM_sov.std()) + ',')
        f.write("%.2f ± %.2f" % (snns.PSSM_sov.mean(),snns.PSSM_sov.std()) + ',')
        f.write("%.2f ± %.2f" % (keras.PSSM_sov.mean(),keras.PSSM_sov.std()) + '\n')
```

```python
outfile='/cluster/gjb_lab/2472402/results/26Aug_CV_summary.csv'
expt_name='25Aug'
with open(outfile, 'w+') as f:
    # write header
    f.write('fold,SNNS_HMM_acc,Keras_HMM_acc,SNNS_PSSM_acc,Keras_PSSM_acc,SNNS_HMM_sov,Keras_HMM_sov,SNNS_PSSM_sov,Keras_PSSM_sov\n')
    col1sum = []
    col2sum = []
    col3sum = []
    col4sum = []
    col5sum = []
    col6sum = []
    col7sum = []
    col8sum = []
    for i in range(1,9):
        
        if i==8:
            col1 = pd.concat(col1sum)
            col2 = pd.concat(col2sum)
            col3 = pd.concat(col3sum)
            col4 = pd.concat(col4sum)
            col5 = pd.concat(col5sum)
            col6 = pd.concat(col6sum) 
            col7 = pd.concat(col7sum)
            col8 = pd.concat(col8sum)
            
            f.write("%d," % 8)
            f.write("%.2f ± %.2f" % (col1.mean(),col1.std()) + ',')
            f.write("%.2f ± %.2f" % (col2.mean(),col2.std()) + ',')
            f.write("%.2f ± %.2f" % (col3.mean(),col3.std()) + ',')
            f.write("%.2f ± %.2f" % (col4.mean(),col4.std()) + ',')
            f.write("%.2f ± %.2f" % (col5.mean(),col5.std()) + ',')
            f.write("%.2f ± %.2f" % (col6.mean(),col6.std()) + ',')
            f.write("%.2f ± %.2f" % (col7.mean(),col7.std()) + ',')
            f.write("%.2f ± %.2f" % (col8.mean(),col8.std()) + '\n')
        else:

            keras_path="/cluster/gjb_lab/2472402/results/keras/%s/cv%d_scores.csv" % (expt_name,i)
            snns_path="/cluster/gjb_lab/2472402/results/snns/%s/cv%d_scores.csv" % (expt_name,i)
            snns=pd.read_csv(snns_path).set_index('seqID')
            keras=pd.read_csv(keras_path).set_index('seqID')

            col1 = snns.HMM_acc
            col1sum.append(col1)
            col2 = keras.HMM_acc
            col2sum.append(col2)
            col3 = snns.PSSM_acc
            col3sum.append(col3)
            col4 = keras.PSSM_acc
            col4sum.append(col4)
            col5 = snns.HMM_sov
            col5sum.append(col5)
            col6 = keras.HMM_sov
            col6sum.append(col6)
            col7 = snns.PSSM_sov
            col7sum.append(col7)
            col8 = keras.PSSM_sov
            col8sum.append(col8)

            f.write("%d," % i)
            f.write("%.2f ± %.2f" % (snns.HMM_acc.mean(),snns.HMM_acc.std()) + ',')
            f.write("%.2f ± %.2f" % (keras.HMM_acc.mean(),keras.HMM_acc.std()) + ',')
            f.write("%.2f ± %.2f" % (snns.PSSM_acc.mean(),snns.PSSM_acc.std()) + ',')
            f.write("%.2f ± %.2f" % (keras.PSSM_acc.mean(),keras.PSSM_acc.std()) + ',')
            f.write("%.2f ± %.2f" % (snns.HMM_sov.mean(),snns.HMM_sov.std()) + ',')
            f.write("%.2f ± %.2f" % (keras.HMM_sov.mean(),keras.HMM_sov.std()) + ',')
            f.write("%.2f ± %.2f" % (snns.PSSM_sov.mean(),snns.PSSM_sov.std()) + ',')
            f.write("%.2f ± %.2f" % (keras.PSSM_sov.mean(),keras.PSSM_sov.std()) + '\n')
```

# Run describe() on every row of every fold of cross validation 

```python
import os

pd.options.display.float_format = "{:.3f}".format
outfile='/cluster/gjb_lab/2472402/results/25Aug_CV_details.txt'
expt_name='25Aug'

keys='SNNS HMM acc,Keras HMM acc,SNNS PSSM acc,Keras PSSM acc,SNNS HMM sov,Keras HMM sov,SNNS PSSM sov,Keras PSSM sov'.replace(' ','_').split(',')

with open(outfile, 'w+') as f:
    for i in range(1,8):
        keras_path="/cluster/gjb_lab/2472402/results/keras/%s/cv%d_scores.csv" % (expt_name,i)
        assert os.path.exists(keras_path)
        snns_path="/cluster/gjb_lab/2472402/results/snns/%s/cv%d_scores.csv" % (expt_name,i)
        assert os.path.exists(snns_path)
        snns=pd.read_csv(snns_path).set_index('seqID')
        assert snns.shape[0]!=0
        keras=pd.read_csv(keras_path).set_index('seqID')
        assert keras.shape[0]!=0
        f.write("Cross-validation fold %d\n" % i)
        df1=snns.describe()
        df2=keras.describe()
        combined_df=pd.concat([df1.HMM_acc, df2.HMM_acc, df1.PSSM_acc, df2.PSSM_acc, df1.HMM_sov, df2.HMM_sov, df1.PSSM_sov, df2.PSSM_sov],axis=1,
         keys=keys)
        combined_df.to_csv(f)
```
